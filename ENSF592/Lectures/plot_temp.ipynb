```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
# Load data with pandas, parse datetime in first column
df = pd.read_csv('temp.csv', 
                header=None, 
                names=['date', 'temp'], 
                skiprows=4, 
                parse_dates=[0,], 
                date_parser=pd.to_datetime)

# Check dtypes to make sure df['date'] is of type datetime
print(df.dtypes)

# Create and format plot
plt.plot(df['date'],df['temp'])
plt.xticks(rotation='vertical')
plt.grid()
```
