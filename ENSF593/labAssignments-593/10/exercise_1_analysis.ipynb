```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sorting_data = pd.read_csv('./ComplexityAnalysis/log.txt', sep=' ')
# sorting_data.loc[sorting_data['algorithm'] == 'insert'].groupby(by=)
```

```python
# for order in sorting_data.groupby('order):
#     plt.subplot

# g = sns.FacetGrid(sorting_data, col="order", hue="algorithm")
# g.map(plt.scatter, "n", "time")
# g.set(yscale="log")
order = 'descending'
plot= sns.lineplot(x='n', y='time', hue='algorithm', data=sorting_data.loc[sorting_data['order'] == order], ci=False)
plot.set(yscale="log", xscale='log', ylim=(0,10**13))
plot.set_title(order)
fig = plot.get_figure()
fig.savefig("output.png")
```
